```python

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from shapely import wkt

# Step 1: Read and Combine the School Data
# --- NOTE: Replace 'schools_file1.csv' and 'schools_file2.csv' with your actual file names.
print("Reading and merging school datasets...")
schools_df1 = pd.read_csv(r"C:\Users\yacin\Downloads\dv402-SchoolLocations2025.csv")
schools_df2 = pd.read_csv(r"C:\Users\yacin\Downloads\dv403-AllSchoolsEnrolments-2025.csv")


# Merge the two school DataFrames using a common key like 'School_No' or 'School_Name'
schools_combined = pd.merge(schools_df1, schools_df2, on='School_Name', how='inner')
print("School datasets merged successfully.")

# Step 2: Convert the combined Schools DataFrame to a GeoDataFrame
# The school location is in the 'Y' (latitude) and 'X' (longitude) columns.
print("Converting schools data to GeoDataFrame...")
schools_combined['geometry'] = [Point(xy) for xy in zip(schools_combined['X'], schools_combined['Y'])]
schools_geo = gpd.GeoDataFrame(schools_combined, geometry='geometry', crs="EPSG:4326")
print("Schools DataFrame converted to GeoDataFrame.")
```

```python
import plotly.graph_objects as go
# Create a Plotly Figure object
fig = go.Figure()

# Plot the transport stops first (base layer)
fig.add_trace(go.Scattermapbox(
    lat=stops.geometry.y,
    lon=stops.geometry.x,
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=8,
        color='blue'
    ),
    name='Transport Stops'
))

# Add the schools on top of the stops
fig.add_trace(go.Scattermapbox(
    lat=schools_geo.geometry.y,
    lon=schools_geo.geometry.x,
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=8,
        color='green'
    ),
    name='Schools'
))

# Add the new housing (permits) as the top layer
fig.add_trace(go.Choroplethmapbox(
    geojson=permits_geo.__geo_interface__,
    locations=permits_geo.index,
    color='red',
    marker_line_width=1,
    opacity=0.6,
    name='New Housing'
))

# Set the map's layout and style
fig.update_layout(
    mapbox_style="carto-positron",  # You can choose other styles like 'open-street-map', 'stamen-terrain', etc.
    mapbox_zoom=12,
    mapbox_center={"lat": stops.geometry.y.mean(), "lon": stops.geometry.x.mean()},
    title='New Housing, Transport Stops, and Schools'
)

# Display the map
fig.show()
```
